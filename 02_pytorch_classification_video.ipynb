# Classification in Machine Learning

Classification is a fundamental task in machine learning, involving the assignment of labels to instances based on their features. This guide delves into the three primary types of classification: **Binary**, **Multiclass**, and **Multilabel**. We'll explore each from basic concepts to advanced methodologies, providing a thorough understanding of their applications and techniques.

## 1. Binary Classification

### Definition

Binary classification involves categorizing instances into one of two distinct classes. It's the simplest form of classification, where the model predicts a label of either 0 or 1, true or false, positive or negative.

### Examples

- **Spam Detection**: Classifying emails as 'spam' or 'not spam'.
- **Medical Diagnosis**: Determining the presence or absence of a disease.
- **Sentiment Analysis**: Identifying text sentiment as 'positive' or 'negative'.

### Common Algorithms

- **Logistic Regression**: Models the probability of a binary outcome based on one or more predictor variables.
- **Support Vector Machines (SVM)**: Finds the optimal hyperplane that separates the two classes.
- **Decision Trees**: Splits data into subsets based on feature values to make predictions.
- **Naive Bayes**: Applies Bayes' theorem with the assumption of feature independence.

### Evaluation Metrics

- **Accuracy**: Proportion of correctly classified instances.
- **Precision**: Proportion of true positive predictions among all positive predictions.
- **Recall (Sensitivity)**: Proportion of true positive predictions among all actual positives.
- **F1 Score**: Harmonic mean of precision and recall.
- **ROC-AUC**: Area under the Receiver Operating Characteristic curve, representing the trade-off between true positive rate and false positive rate.

## 2. Multiclass Classification

### Definition

Multiclass classification deals with problems where instances are categorized into one of three or more classes. Unlike binary classification, each instance belongs exclusively to a single class.

### Examples

- **Image Recognition**: Classifying images into categories like 'cat', 'dog', 'bird', etc.
- **Handwritten Digit Recognition**: Identifying digits from 0 to 9.
- **Language Identification**: Determining the language of a given text.

### Strategies

- **One-vs-Rest (OvR)**: Trains one classifier per class, with the samples of that class as positive samples and all others as negatives. During prediction, the classifier with the highest confidence score determines the class. :contentReference[oaicite:0]{index=0}
- **One-vs-One (OvO)**: Trains a classifier for every pair of classes. For prediction, a voting scheme is applied where each classifier votes for one class, and the class with the most votes is selected. :contentReference[oaicite:1]{index=1}

### Common Algorithms

- **Softmax Regression**: Generalization of logistic regression for multiple classes.
- **Decision Trees and Random Forests**: Naturally handle multiple classes by splitting data based on feature values.
- **Neural Networks**: Utilize architectures like feedforward networks with a softmax output layer to handle multiple classes.

### Evaluation Metrics

- **Confusion Matrix**: A table layout that allows visualization of the performance of an algorithm.
- **Precision, Recall, and F1 Score**: Can be averaged using macro (unweighted mean) or micro (weighted mean) methods across all classes.
- **Accuracy**: Overall proportion of correctly classified instances.

## 3. Multilabel Classification

### Definition

In multilabel classification, each instance can be assigned multiple labels simultaneously. This is different from multiclass classification, where each instance is associated with only one class.

### Examples

- **Text Categorization**: Assigning multiple topics to a document.
- **Image Tagging**: Associating multiple tags (e.g., 'beach', 'sunset', 'vacation') with a single image.
- **Music Genre Classification**: Identifying multiple genres that a song may belong to.

### Problem Transformation Methods

- **Binary Relevance**: Treats each label as a separate binary classification problem, training one classifier per label independently. :contentReference[oaicite:2]{index=2}
- **Classifier Chains**: Extends binary relevance by chaining classifiers so that each classifier deals with the binary classification problem associated with each label, taking into account the predictions of previous classifiers in the chain. :contentReference[oaicite:3]{index=3}
- **Label Powerset**: Transforms the problem into a multiclass classification problem by considering each unique combination of labels as a single class. :contentReference[oaicite:4]{index=4}

### Adapted Algorithms

- **Multilabel k-Nearest Neighbors (ML-kNN)**: Extends k-NN to handle multilabel data by considering the labels of the nearest neighbors. :contentReference[oaicite:5]{index=5}
- **Multilabel Decision Trees**: Modifies decision tree algorithms to handle multiple labels by adjusting the splitting criteria and leaf node predictions. :contentReference[oaicite:6]{index=6}
- **Neural Networks**: Utilize architectures that output a vector of probabilities, one for each label, often employing a sigmoid activation function for each output node.

### Evaluation Metrics

- **Hamming Loss**: Fraction of labels that are incorrectly predicted.
- **Subset Accuracy**: Proportion of instances where the entire set of predicted labels matches the true set of labels.
- **Precision, Recall, and F1 Score**: Can be computed for each label and then averaged.
- **Ranking Loss**: Measures the average fraction of label pairs that are incorrectly ordered.


### Deep Learning Approaches

Deep learning models, especially neural networks, have been adapted for complex classification tasks:

- **Convolutional Neural Networks (CNNs)**: Effective for image data, capable of handling multiclass and multilabel classification by adjusting the output layer and loss functions.
- **Recurrent Neural Networks (RNNs)**: Suitable for sequential data like text, can be adapted for multilabel text classification tasks.

# Classification Inputs and Outputs

## **1. Overview**
This diagram explains how an image classification model processes inputs and generates outputs. The goal is to classify food images into categories like **Sushi , Steak , and Pizza **.

- **Actual Output**: The correct label assigned to an image.
- **Predicted Output**: The classification made by the machine learning model.

---

## **2. Inputs (Image Representation)**
### **Image Dimensions and Color Channels**
Each image is resized to **224  224 pixels** with **3 color channels (RGB)**:
- **Width (W) = 224**
- **Height (H) = 224**
- **Channels (C) = 3** (Red, Green, and Blue)

### **Numerical Encoding (Pixel Values)**
The image is converted into a matrix of pixel values, **normalized between 0 and 1**:
```plaintext
[ [0.31, 0.62, 0.44, ...],
  [0.92, 0.03, 0.27, ...],
  [0.25, 0.78, 0.07, ...] ]
```
Each pixel originally has values in the range **[0, 255]** (since RGB values go from 0 to 255). Normalization is done as follows:
```plaintext
Normalized Value = Pixel Value / 255
```
For example, if the original pixel RGB value was **(80, 160, 112)**:
```plaintext
R = 80 / 255 = 0.31
G = 160 / 255 = 0.62
B = 112 / 255 = 0.44
```

---

## **3. Machine Learning Model**
- The **numerical data** is fed into a **Machine Learning Algorithm**, such as a **Convolutional Neural Network (CNN)**.
- CNN extracts patterns and features from the pixel values.
- The model assigns **probabilities to each class** (e.g., Sushi, Steak, Pizza).

---

## **4. Outputs (Predictions)**
The model generates a probability distribution for different classes:
```plaintext
[0.97, 0.00, 0.03]  Sushi 
[0.81, 0.14, 0.05]  Incorrect (should be Steak )
[0.03, 0.07, 0.90]  Pizza 
```
- The **highest probability** determines the predicted category.
- A prediction is **correct ()** if it matches the actual label; otherwise, it's incorrect ().

---

## **5. Key Takeaways**
- **Image Classification Workflow**:
  1. Convert images into numerical form (RGB pixel values).
  2. Feed the processed data into a machine learning model.
  3. The model generates probability scores for each class.
  4. The highest probability determines the predicted category.
  5. Compare predictions with actual labels to evaluate accuracy.

- **Model Learning**:
  - The model improves by training on large datasets of labeled images.
  - It learns to recognize patterns by looking at many examples.

## **1. Overview**
This diagram explains how **image classification models** handle input and output shapes. The key elements include:
- **Input Shape**: How image data is structured before entering the model.
- **Machine Learning Algorithm**: Processes the input and generates predictions.
- **Output Shape**: The final probability distribution for classification.

---

## **2. Input Representation (Tensor Shape)**
### **Image Dimensions and Color Channels**
Each image has the following properties:
- **Width = 224 pixels**
- **Height = 224 pixels**
- **Channels = 3 (RGB)**

### **Numerical Encoding of Pixels**
Each pixel is converted into a **normalized** value between 0 and 1:
```plaintext
[ [0.31, 0.62, 0.44, ...],
  [0.92, 0.03, 0.27, ...],
  [0.25, 0.78, 0.07, ...] ]
```

### **Tensor Representation**
Images are represented as tensors in machine learning models:
```plaintext
[batch_size, color_channels, width, height]
```
#### **Possible Shapes**
- **Single Image (Flexible Batch Size)**:
  ```plaintext
  Shape = [None, 3, 224, 224]
  ```
  (`None` means batch size is unspecified, allowing flexibility in training.)

- **Batch of 32 Images (Common Practice in Training)**:
  ```plaintext
  Shape = [32, 3, 224, 224]
  ```
  (`32` is a common batch size used in deep learning.)

---

## **3. Machine Learning Model**
- The input tensor is passed into a **Machine Learning Algorithm**.
- The model extracts patterns and features from pixel values.
- It outputs a probability distribution over possible categories.

---

## **4. Output Representation**
### **Prediction Probabilities**
The model produces a probability vector representing confidence levels for each class:
```plaintext
[0.97, 0.00, 0.03]  Sushi 
```
- The highest probability determines the predicted category.

### **Output Shape**
The output is a 1D tensor where each element corresponds to a class probability:
```plaintext
Shape = [3]
```
(Here, the model predicts 3 different classes: **Sushi , Steak , and Pizza **.)

---

## **5. Key Takeaways**
- **Input Shape**: `[batch_size, 3, 224, 224]` (batch size can vary, commonly 32).
- **Output Shape**: `[3]` (one probability per class).
- **Tensor Representation**: Enables the model to process multiple images efficiently.
- **Batch Processing**: Helps in efficient computation during training.
- **Model Learning**: The model refines its classification by learning from a dataset of labeled images.

---

## **Conclusion**
This visualization clarifies how input images are structured into tensors and how deep learning models predict class probabilities. The flexibility of batch size helps in both training and inference, making the model more efficient and scalable.

# Architecture of a Classification Model

## **Overview**
This document explains the typical architecture of a classification model, comparing **binary classification** and **multiclass classification** approaches, including key hyperparameters and a sample PyTorch model.

---

## **Hyperparameter Breakdown**

| **Hyperparameter**           | **Binary Classification** | **Multiclass Classification** |
|-----------------------------|--------------------------|------------------------------|
| **Input layer shape (_in_features_)** | Matches the number of input features (e.g., 5 for age, sex, height, etc.) | Same as binary classification |
| **Hidden layer(s)**          | Problem-specific, minimum 1, maximum unlimited | Same as binary classification |
| **Neurons per hidden layer** | Typically between 10 to 512 | Same as binary classification |
| **Output layer shape (_out_features_)** | 1 (for binary classification) | 1 per class (e.g., 3 for food types) |
| **Hidden layer activation**  | Usually **ReLU** (Rectified Linear Unit), but others exist | Same as binary classification |
| **Output activation**        | **Sigmoid** (`torch.sigmoid` in PyTorch) | **Softmax** (`torch.softmax` in PyTorch) |
| **Loss function**            | **Binary Cross Entropy** (`torch.nn.BCELoss`) | **Cross Entropy** (`torch.nn.CrossEntropyLoss`) |
| **Optimizer**                | **SGD** (`stochastic gradient descent`) or **Adam** | Same as binary classification |

---


---

## **Workflow Overview**
1. **Image Input:** Images (e.g., sushi, steak, pizza) are passed as input.
2. **Feature Extraction:** The model extracts features and passes them through layers.
3. **Activation & Loss Function:** Activations (ReLU/Sigmoid/Softmax) help in decision-making.
4. **Prediction Output:** The model predicts probabilities for different classes (e.g., sushi , steak , pizza ).

This architecture serves as a **foundation** for building classification models in deep learning!

---

### **Key Takeaways:**
- **Binary classification** uses a **single output node** with Sigmoid activation.
- **Multiclass classification** uses multiple output nodes with **Softmax**.
- **Cross Entropy Loss** is used for **multiclass problems**, while **Binary Cross Entropy** is for **binary problems**.

# 2. Building a model

Let's build a model to classify our blue and red dots.

To do so, we want to:
1. Setup device agnoistic code so our code will run on an accelerator (GPU) if there is one
2. Construct a model (by subclassing `nn.Module`.)
3. Define a loss funtion and a optimizer.
4. Create a training and test loop.


````python
# Import PyTorch and nn
import torch
from torch import nn

# Make device agnostic code
device = "cuda" if torch.cuda.is_available() else "cpu"
device
````

Now we've setup device agnostic code, let's create a model that:
1. Subclasses `nn.Module` (almost all models in PyTorch subclass `nn.Module`)
2. Create two `nn.Linear() Layers` that are caple of handling the shapes of our data.
3. Defines a `forward()` method  that outlines the forward pass (or forward computation) of the model.
4. Instantiate an instance of our model class and set it to the target device.

````python
X_train.shape
````

````python
y_train[:5]
````

````python
# 1. Construct a model that subclasses nn.Module
class CircleModelV0(nn.Module):
  def __init__(self):
    super().__init__()
    # 2. Create two nn.Linear layer capable of handling the shapes of our data
    self.layer_1 = nn.Linear(in_features=2, out_features=5) # takes in 2 features and upscales to 5 features
    self.layer_2 = nn.Linear(in_features=5, out_features=1) # takes in 5 features from previous layer and outputs a single feature (same shape as y)

    # 3. Define a forward() method that outlines the forward pass
    def forward(self, x):
      return self.layer_2(self.layer_1(x)) # x -> layer_1 -> layer_2 -> output

# 4. Instantiate an unstance of our model class and send it to the target device
model_0 = CircleModelV0().to(device)
model_0
````
